## Reobtain the data and generators

```
# Read from the csv 
df_train = pd.read_csv('drive/My Drive/data/malaria_train_combined.csv')
df_val = pd.read_csv('drive/My Drive/data/malaria_val_combined.csv')
df_test = pd.read_csv('drive/My Drive/data/malaria_test_combined.csv')

df_train['labels'] = df_train['labels'].astype(str)
df_val['labels'] = df_val['labels'].astype(str)
df_test['labels'] = df_test['labels'].astype(str)
```

```
from keras.applications.xception import preprocess_input

# Create generators manually
train_datagen = ImageDataGenerator(
     preprocessing_function=lambda x: preprocess_input(x),
     shear_range=0.2,
     zoom_range=0.2,
     #width_shift_range=0.2,
     #height_shift_range=0.2,
     horizontal_flip=True)
  
val_datagen = ImageDataGenerator(preprocessing_function=lambda x: preprocess_input(x))
```

```
# Get the generators
train_gen, valid_gen, test_gen = fit_into_generators(df_train,df_val,df_test,train_datagen, val_datagen,batch_size=32,shape=(299,299))
```

```
from keras import layers, optimizers
```

```
# Reload finetuned-inceptv3 model
model = make_finetuned_xcept()

model.load_weights('/content/drive/My Drive/weights/finetuned_xceptv1_weights_for_predicting_malaria.h5')
model.compile(optimizers.Adam(0.01), loss = "categorical_crossentropy", metrics=['accuracy'])
```

```
model.evaluate_generator(test_gen)
```

## Visualisation of 'Explainability' of finetuned xceptv1 model 

```
import imp
import innvestigate
import innvestigate.utils as iutils

imgnetutils = imp.load_source("utils_imagenet", "/content/drive/My Drive/src/utils_imagenet.py")
```

```
# Define methods
methods = [
    # NAME                    OPT.PARAMS                POSTPROC FXN                TITLE
    # Function
    #("gradient",              {"postprocess": "abs"},   imgnetutils.graymap,       "Gradient"),
    
    ("smoothgrad",            {"postprocess": "square"},imgnetutils.graymap,       "SmoothGrad"),

    ("lrp.sequential_preset_a_flat",{"epsilon": 1},     imgnetutils.heatmap,       "LRP-PresetAFlat")
]
```

```
from keras import Model

# This is to remove 'model' in the layer for innvestigate to work
x = model.layers[0].layers[-1].get_output_at(0)

for layer in model.layers[1:]:
  x = layer(x)
model_new = Model(inputs=model.layers[0].inputs, outputs=x)
```

```
# Create model without trailing softmax
model_wo_softmax = iutils.keras.graph.model_wo_softmax(model_new)

# Create analyzers.
analyzers = []
for method in methods:
  analyzer = innvestigate.create_analyzer(method[0],        # analysis method identifier
                                          model_wo_softmax, # model without softmax output
                                          **method[1])      # optional analysis parameters
  analyzers.append(analyzer)
```

```
analyzer_names = ['Smooth Gradient', 'LRP Preset A Flat']
```

### True positive samples 

```
# 4 random true positive samples
sample_indices = [np.random.choice(len(tp[0])) for i in range(4)]
print('The sample indices for true positives are {}'.format(sample_indices))
```

```
for image in sample_indices:
  show_explanation(tp[:,image,:,:,:],[analyzers[1]],[analyzer_names[1]],[methods[1]])
```

For the true positive samples from the finetuned XceptionV1 model, it seems that the model is associating a small number of pixels in the correct region with the prediction 'Parasitized'. It could be that it is focusing on the relatively darker pixels within the dark region (parasitized), rather than the whole dark region. However, it is also less confident of predicting 'Parasitized' because of the nearby regions.

### True Negative Samples

```
# 4 random true negative samples
sample_indices = [np.random.choice(len(tn[0])) for i in range(4)]
print('The sample indices for true negatives are {}'.format(sample_indices))
```

```
for image in sample_indices:
  show_explanation(tn[:,image,:,:,:],[analyzers[1]],[analyzer_names[1]],[methods[1]])
```

For the true negative samples, an interesting point here is that instead of checking the boundary of the cell to confirm that the cell is 'Unparasitized', a large portion of the interior of the cell is checked, which is indicated by the red pixeled regions in the cell interior for the LRP Preset A heatmap. The model is also less confident of predicting 'Unparasitized' (Blue) because of some parts within the interior as well.

### False Positive

```
# 4 random false positive samples
sample_indices = [np.random.choice(len(fp[0])) for i in range(4)]
print('The sample indices for false positives are {}'.format(sample_indices))
```

```
for image in sample_indices:
  show_explanation(fp[:,image,:,:,:],[analyzers[1]],[analyzer_names[1]],[methods[1]])
```

For the false positives, the model is confused by some of the relatively darker parts in the preprocessed images (Red). Also, it is somewhat uncertain that the parts nearby are indicative of 'Parasitized' (Blue). This means that the model is quite 'close' to predicting the right class 'Unparasitized', just that it is unclear whether the dark region is indicative of 'Parasitized'.

### False Negatives

```
# 4 random false negative samples
sample_indices = [np.random.choice(len(fn[0])) for i in range(4)]
print('The sample indices for false negatives are {}'.format(sample_indices))
```

```
for image in sample_indices:
  show_explanation(fn[:,image,:,:,:],[analyzers[1]],[analyzer_names[1]],[methods[1]])
```

For the case of false negatives, on average, the model is looking at the interior to determine that the cell is 'Unparasitized' when it is in fact not (Red). At the same time, for the 3rd and 4th samples, it seems that the model is almost close to predicting the alternative class 'Parasitized' which is the correct class by looking at the correct parts(Blue).
